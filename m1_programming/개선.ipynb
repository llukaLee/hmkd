```python
import random

class LunchWorldCup:
    def __init__(self):
        self.foodIndex = ["햄버거","탕수육","피자","치킨","자장면","짬뽕","족발","라면","스테이크","파스타","샌드위치","타코","카레",
                          "샐러드","떡볶이","초밥","칼국수","갈비찜","김치찌개","삼겹살","돈가스","볶음밥","불고기","브리또",
                          "김밥","파전","해물찜","갈비탕","감자탕","떡갈비","국밥","찜닭"]
        self.nextList = []
        self.run = True

    def start_game(self):
        print("<<<점심메뉴 이상형 월드컵>>>\n")
        print("총 32개의 메뉴가 있습니다. 중간에 0을 입력하면 종료합니다.")
        print("1. 32강\n2. 16강\n3. 8강")
        rounds = {"32" : 32, "16" : 16, "8" : 8, "1" : 32, "2" : 16, "3" : 8}
        while True:
            select = input()
            if select in rounds:
                self.foodIndex = random.sample(self.foodIndex, rounds.get(select))
                print(f"{rounds.get(select)}강으로 시작합니다.")
                break
            elif select == "0":
                self.run = False
                break
            else:
                print("잘못된 입력입니다. 다시 입력해주세요.")

    def play_game(self):
        while len(self.foodIndex)+len(self.nextList) != 1 and self.run:
            Round = len(self.foodIndex)
            print(f'{Round}강')
            while len(self.foodIndex) != 0 and self.run:
                random.shuffle(self.foodIndex)
                a = self.foodIndex.pop()
                b = self.foodIndex.pop()
                print(a, b)
                while 1:
                    choose = int(input("1번 2번 중에 선택하세요!!\n>"))
                    if choose == 1:
                        self.nextList.append(a)
                        break
                    elif choose == 2:
                        self.nextList.append(b)
                        break
                    elif choose == 0:
                        self.run = False
                        break
                    else:
                        print('다시 입력해 주세요')
            self.foodIndex = self.nextList
            self.nextList = []

    def end_game(self):
        if self.run == False:
            print("종료되었습니다.")
        else:
            print(f"우승은 {self.foodIndex}입니다!!!")

if __name__ == '__main__':
    lunch_wc = LunchWorldCup()
    lunch_wc.start_game()
    lunch_wc.play_game()
    lunch_wc.end_game()
```
